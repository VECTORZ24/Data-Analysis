```python
def remove_page_duplicates(raw_data, target_column):
    def dedupe_consecutive(s):
        parts = s.split("-")
        result = [parts[0]]
        for p in parts[1:]:
            if p != result[-1]:
                result.append(p)
        return "-".join(result)

    data = raw_data.copy()
    data[target_column] = data[target_column].apply(dedupe_consecutive)
    return data
```

```python
data=remove_page_duplicates(raw_data,'user_journey')
data
```
